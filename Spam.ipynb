```python
from sklearn.feature_extraction.text import TfidfVectorizer
# from sklearn.feature_extraction.text import CountVectorizer
# cv=CountVectorizer(max_features=2200)
vectorizer = TfidfVectorizer(ngram_range=(1, 2),min_df=2,max_features=1500)
train_X = vectorizer.fit_transform(train_df['text'])
valid_X = vectorizer.fit_transform(valid_df['text'])
test_X = vectorizer.fit_transform(test_df['text'])
y_train = train_df['label']
y_valid = valid_df['label']
y_test = test_df['label']

print(train_X.shape)
print(valid_X.shape)
print(test_X.shape)
print(y_train.shape)
print(y_valid.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
```

```python
clf = LogisticRegression(random_state=0).fit(train_X, y_train)
y_pred_clf=clf.predict(valid_X)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
accuracy_score(y_valid,y_pred_clf)
```

```python
clf = MultinomialNB()
clf.fit(train_X, y_train)
y_pred_clf=clf.predict(valid_X)
accuracy_score(y_valid,y_pred_clf)
```

```python
clf = DecisionTreeClassifier(random_state=0).fit(train_X, y_train)
y_pred_clf=clf.predict(valid_X)
accuracy_score(y_valid,y_pred_clf)

```

```python
clf = LinearSVC(random_state=0).fit(train_X, y_train)
y_pred_clf=clf.predict(valid_X)
accuracy_score(y_valid,y_pred_clf)
```
